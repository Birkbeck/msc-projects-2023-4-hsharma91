```python
import matplotlib.pyplot as plt
from PIL import Image

# Load the images
img_paths = [
    'C:/Users/hshar/Msc Project/Project Code/Temporary/Feature selection/Adversarial Attack.png',
    'C:/Users/hshar/Msc Project/Project Code/Temporary/Feature selection/Cryptojacking.png',
    'C:/Users/hshar/Msc Project/Project Code/Temporary/Feature selection/Data Poisoning.png',
    'C:/Users/hshar/Msc Project/Project Code/Temporary/Feature selection/Deepfake.png',
    'C:/Users/hshar/Msc Project/Project Code/Temporary/Feature selection/IoT Device Attack.png',
    'C:/Users/hshar/Msc Project/Project Code/Temporary/Feature selection/Ransomware.png',
    'C:/Users/hshar/Msc Project/Project Code/Temporary/Feature selection/WannaCry Ransomware.png'
]

images = [Image.open(img_path) for img_path in img_paths]

# Adjusting the layout to maximize image size for visibility and make text clear
fig, axs = plt.subplots(4, 2, figsize=(40,50))  # Maximize figure size to fit two images per row

# Set the title with appropriate padding
fig.suptitle('Random Forest Based Feature Selection', fontsize=80)

# Remove the extra subplot (since we have only 7 images)
axs[3, 1].axis('off')

# Plot each image in the subplot
for i, ax in enumerate(axs.flat[:-1]):  # Iterate over the 7 available axes
    ax.imshow(images[i])
    ax.axis('off')  # Hide axes

# Adjust layout to make images as large as possible for better visibility
plt.subplots_adjust(wspace=0.01, hspace=0.0001, top=0.95)

# Display the updated collage
plt.show()
```
